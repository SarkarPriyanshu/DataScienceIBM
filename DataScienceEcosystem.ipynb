```python
# This will convert 200 minutes to hours by substracting by 60
# logic....
# 200-60
# 140-60
# 80-60

hours = 0
def mintohours(minutes,hours):
    if minutes <= 60:
        return f'{hours} hr:{minutes} min'
    hours += 1
    return mintohours(minutes-60,hours)

print(mintohours(800,hours))
```
